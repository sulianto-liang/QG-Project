## Import the library needed

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load the Paid Transaction sheet into data frame

```python
df_paidtrx = pd.read_excel('Homework-DataFrame-Dataset.xlsx','Paid-Transaction')
```

```python
df_paidtrx
```

## Combine First Name & Last Name into Full Name

```python
FullName = df_paidtrx["First Name"] + " " + df_paidtrx["Last Name"]
```

## Create new column "Full Name"

```python
df_paidtrx.insert(4,"Full Name",FullName)
```

## Set the "Full Name" column into title case

```python
df_paidtrx["Full Name"] = df_paidtrx["Full Name"].str.title()
```

```python
df_paidtrx
```

## Calculate GMV

```python
GMV = df_paidtrx['Transaction Amount'] - df_paidtrx['Seller Discount'] + df_paidtrx['Delivery Fee']
```

## Create new column "GMV"

```python
df_paidtrx.insert(9,"GMV",GMV)
```

```python
df_paidtrx
```

## Determine "Group" conditions based on "Meta Category" & create a new column "Group"

```python
conditions = [
    df_paidtrx['Meta Category'].isin(['Home/ Food','Sports/ Hobi/ Otomotif','Fashion','Beauty/ Health']),
    df_paidtrx['Meta Category'].isin(['Gadget/ Komputer','Elektronik','Service/ Mokado','Babies/ Kids'])
]

results = ['Group1','Group2']

df_paidtrx['Group'] = np.select(conditions,results,default='')
```

```python
df_paidtrx
```

## Load the Seller sheet into data frame

```python
df_seller = pd.read_excel('Homework-DataFrame-Dataset.xlsx','Seller')
```

```python
df_seller
```

## Merge seller data frame into paid transaction data frame

```python
df_paidtrx = df_paidtrx.merge(df_seller, how="left", left_on=['Order Number'], right_on=['Order Number'])
```

```python
df_paidtrx
```

## Change "Paid Date" data type into datetime

```python
df_paidtrx['Paid Date'] = df_paidtrx['Paid Date'].astype(str)
df_paidtrx['Paid Date'] = df_paidtrx['Paid Date'].str[:4] + "-" + df_paidtrx['Paid Date'].str[4:6] + "-" + df_paidtrx['Paid Date'].str[6:8]
df_paidtrx['Paid Date'] = pd.to_datetime(df_paidtrx['Paid Date'])
```

## Add new column "MonthYear"

```python
df_paidtrx['MonthYear'] = df_paidtrx['Paid Date'].dt.month_name() + " " + df_paidtrx['Paid Date'].dt.year.astype(str)
```

## Aggregating Seller by Month & GMV

```python
df_seller_mon_gmv = df_paidtrx.groupby(['MonthYear','Seller'],as_index=False).agg({'GMV':'sum'})
```

```python
df_seller_mon_gmv
```

## Find seller with the highest GMV on August 2017

```python
df_seller_august2017 = df_seller_mon_gmv[df_seller_mon_gmv["MonthYear"]=="August 2017"]
df_seller_august2017[df_seller_august2017["GMV"]==df_seller_august2017["GMV"].max()]
```

## Aggregating Seller by Month & Count of Trx for Meta Category = "Fashion"

```python
df_seller_mon_fashion = df_paidtrx[df_paidtrx["Meta Category"]=="Fashion"].groupby(['MonthYear','Seller'],as_index=False).agg({'Order Number':'count'})
df_seller_mon_fashion.rename(columns={'Order Number':'Count of Trx'},inplace=True)
```

## Find seller with the highest Count of Trx on September 2017 for Meta Category "Fashion"

```python
df_seller_fashion_september2017 = df_seller_mon_fashion[df_seller_mon_fashion["MonthYear"]=="September 2017"]
df_seller_fashion_september2017[df_seller_fashion_september2017["Count of Trx"]==df_seller_fashion_september2017["Count of Trx"].max()]
```
